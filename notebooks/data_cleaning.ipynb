```python
import pandas as pd

file_path = r"E:\Projects\Marketing Analytics\data\raw\marketing_campaign.csv"

df = pd.read_csv(file_path)
```

```python
df.info()
```

```python
# Lowering all the column names

df.columns = df.columns.str.lower()

df.head()
```

```python
df.columns = (
    df.columns
      .str.lower()
      .str.replace(r'[^a-z0-9]+', '_', regex=True)
      .str.strip('_')
)

df.info()
```

```python
# Converting date to datetime

df['date'] = df['date'] = pd.to_datetime(df['date'])

df.info()
```

```python
# Removing $ from columns 

cost_cols = [
    'cost_per_facebook_ad',
    'cost_per_adwords_ad',
    'facebook_cost_per_click_ad_cost_clicks',
    'adwords_cost_per_click_ad_cost_clicks'
]

for col in cost_cols:
    df[col] = (
        df[col]
        .str.replace('$', '', regex=False)
        .astype(float)
    )
```

```python
df.head()
```

```python
# Removing % sign from columns

pct_cols = [
    'facebook_click_through_rate_clicks_view',
    'facebook_conversion_rate_conversions_clicks',
    'adwords_click_through_rate_clicks_view',
    'adwords_conversion_rate_conversions_click'
]

for col in pct_cols:
    df[col] = (
        df[col]
        .str.replace('%', '', regex=False)
        .astype(float)
        / 100
    )
```

```python
df.head()
```

```python
# checking missing values , unique values and duplicates 

eda_check = pd.DataFrame({
    'missing_%': df.isna().mean() * 100,
    'unique_values': df.nunique()
})

eda_check.loc['__overall__', 'duplicate_rows'] = df.duplicated().sum()

eda_check
```

```python
df.to_csv("marketing_cleaned.csv", index=False)
```
